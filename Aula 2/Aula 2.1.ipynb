```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
df = pd.read_json('dados.txt')
df.head()
```

```python
df.rename({'Film':'filme',
            'Year':'ano',
            'Genre':'genero',
            'Worldwide_Gross_M':'bilheteria',
            'Budget_M':'orcamento'}, axis = 1,  inplace=True)
df.head()
```

```python
filmes = df.filme
orcamentos = df.orcamento

plt.figure(figsize=(120, 10))
plt.bar(filmes,orcamentos)
plt.xlabel('Filmes')
plt.ylabel('Orçamento')
plt.title('Gráfico de orçamento por filme')
plt.show()
```

```python
plt.figure(figsize=(15,8))
plt.title('Orçamento de todos os filmes')
sns.barplot(y='filme', x='orcamento', color='red', data=df)
plt.show()
```

```python
ordenado = df.sort_values(by='orcamento', ascending = True)

plt.figure(figsize=(15,8))
plt.title('Orçamento de todos os filmes')
sns.barplot(y='filme', x='orcamento', color='red', data=ordenado)
plt.show()
```

```python
fl = df.sort_values(by='orcamento', ascending = False).head(1)
mo = fl['filme']
print('O  filme com  o  maior orçamento é: {}'.format(mo.values[0]))

fl = df.sort_values(by='orcamento', ascending = True).head(1)
mo = fl['filme']
print('\nO  filme com  o  menor orçamento é: {}'.format(mo.values[0]))
```

```python
plt.figure(figsize=(15,8))
plt.title('Bilheteria de todos os filmes')
sns.barplot(y='filme', x='bilheteria', color='red', data=df)
plt.show()
```

```python
ordenado = df.sort_values(by='bilheteria', ascending = True)

plt.figure(figsize=(15,8))
plt.title('Orçamento de todos os filmes')
sns.barplot(y='filme', x='bilheteria', color='red', data=ordenado)
plt.show()
```

```python
fl = df.sort_values(by='bilheteria', ascending = False).head(1)
mo = fl['filme']
print('O  filme com  o  maior orçamento é: {}'.format(mo.values[0]))

fl = df.sort_values(by='bilheteria', ascending = True).head(1)
mo = fl['filme']
print('\nO  filme com  o  menor orçamento é: {}'.format(mo.values[0]))
```

```python
grupoGenero = df.groupby(by='genero').count()
grupoGenero
```

```python
plt.figure(figsize=(13,8))
plt.title('Filmes por gênero em cada ano')
sns.countplot(x="ano", hue= 'genero', palette = "pastel", edgecolor="1", data=df)
plt.show()
```

```python
print('O ano em que foi feito menos filmes de animação foi em: 2008 e 2011')
```

```python
plt.figure(figsize=(10,8))
plt.title('Média de orçamento por gênero em cada ano')
sns.barplot(x='ano',  y='orcamento', ci = None, hue='genero', palette = "pastel", edgecolor="1", data=df)
plt.show()
```

```python
print('O ano em que os filmes de ação tiveram maior média de orçamento foi: 2007')
print('O ano em que os filmes de animação tiveram maior média de orçamento foi: 2008')
```
