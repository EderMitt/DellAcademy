```python
import pandas as pd
```

```python
df = pd.read_csv('drinks.txt')
```

```python
df.head()
```

```python
df.rename({'country':'pais',
            'beer_servings':'cerveja',
            'spirit_servings':'licor',
            'wine_servings':'vinho',
            'total_litres_of_pure_alcohol':'litros_totais_alcool_puro'}, axis = 1,  inplace=True)
df.head()
```

```python
df
```

```python
descricao = df[['cerveja', 'licor', 'vinho']].describe()
descricao
```

```python
cerveja = df['cerveja']
media_cerveja = "%.2f" % cerveja.mean()
dp_cerveja = "%.2f" % cerveja.std()
print(media_cerveja)

licor = df['licor']
media_licor = "%.2f" % licor.mean()
dp_licor = "%.2f" % licor.std()

vinho = df['vinho']
media_vinho = "%.2f" % vinho.mean()
dp_vinho = "%.2f" % vinho.std()

total_alcool = df['litros_totais_alcool_puro']
media_total_alcool = "%.2f" % total_alcool.mean()
dp_total_alcool = "%.2f" % total_alcool.std()
```

```python
media_total_alcool
```

```python
france = df[df['pais'] == 'France']
licor_france = france.licor[61]
print(licor_france)
```

```python
ord_vinho = df[{'pais', 'vinho'}]
cinco_paises = ord_vinho.sort_values(by='vinho', ascending = False).head()
cinco_paises
```

```python
med_cerveja = df['cerveja'].mean()
paises_cerveja = df[df['cerveja'] > med_cerveja]
paises_cerveja = paises_cerveja.drop(paises_cerveja.columns[2:], axis=1)
paises_cerveja = paises_cerveja.sort_values(by='cerveja', ascending = False)
paises_cerveja
```

```python
brasil = df[df['pais'] == 'Brazil']
cons_brasil = brasil['litros_totais_alcool_puro'][23]
rest_paises = df[df['litros_totais_alcool_puro'] > cons_brasil]
paises = rest_paises.sort_values(by='litros_totais_alcool_puro', ascending = True)
paises = paises[{'pais', 'litros_totais_alcool_puro'}]
paises
```

```python
#Questão 1

print('Qual a média e desvio padrão do consumo de cerveja, licor, vinho e total de álcool?\n')
print(f'\nCerveja\nMédia: {media_cerveja}, Desvio: {dp_cerveja}')
print(f'\nLicor\nMédia: {media_licor}, Desvio: {dp_licor}')
print(f'\nVinho\nMédia: {media_vinho}, Desvio: {dp_vinho}')
print(f'\nTotal de Alcool\nMédia: {media_total_alcool}, Desvio: {dp_total_alcool}')
```
