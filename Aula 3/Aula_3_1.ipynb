```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('dados1.txt')
df.head()
```

```python
df.rename({'city':'cidade',
           'area':'area',
           'rooms':'quarto',
           'bathroom':'banheiro',
           'parking spaces':'estacionamento',
           'floor':'piso',
           'animal':'pet',
           'furniture':'mobiliario',
           'hoa (R$)':'posicao',
           'rent amount (R$)':'aluguel(R$)',
           'property tax (R$)':'iptu(R$)',
           'fire insurance (R$)':'seguro_incendio(R$)',
           'total (R$)':'total (R$)',
           }, axis = 1,  inplace=True)
df.head()
```

```python
df.describe()
```

```python
plt.figure(figsize=(25, 4))
plt.style.use('seaborn-darkgrid')
plt.title('Valores de aluguéis mais comuns')
plt.hist(df['aluguel(R$)'],bins=25)

plt.figure(figsize=(25, 4))
plt.title('Valores de aluguéis mais comuns')
sns.distplot(df['aluguel(R$)'])

plt.figure(figsize=(25, 4))
plt.title('Valores de aluguéis mais comuns')
sns.kdeplot(df['aluguel(R$)'], shade=True)

plt.figure(figsize=(25, 4))
plt.title('Valores de aluguéis mais comuns')
sns.kdeplot(df['aluguel(R$)'],df['area'], shade = True, cmap = "winter_r", cbar=True)
plt.show()
```
