```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style="darkgrid")
```

```python
tips = sns.load_dataset('mpg')
tips.head()
```

```python
plt.figure(figsize=(8,5))
plt.title('Relação entre aceleração e consumo de combustível')
sns.scatterplot(x="acceleration", y="mpg", data=tips)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.title('Relação entre aceleração e consumo de combustível')
sns.regplot(x="acceleration", y="mpg", data=tips)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.title('Relação entre aceleração e consumo de combustível por país')
sns.scatterplot(x="acceleration", y="mpg", hue = 'origin', data=tips)
plt.show()
```

```python
sns.relplot(x="acceleration", y="mpg", col = 'origin', data=tips)
```

```python
sns.lmplot(x="acceleration", y="mpg", hue = 'origin', data=tips)
```

```python
plt.figure(figsize=(8,5))
plt.title('Relação entre aceleração e consumo de combustível por país')
sns.scatterplot(x="acceleration", y="mpg", hue='origin', style ='origin', size = 'origin', sizes=(100, 50), data=tips)
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.title('Relação entre aceleração e consumo de combustível por ano do veículo')
sns.scatterplot(x="acceleration", y="mpg", hue='model_year', style ='model_year', sizes=(100), data=tips)
plt.show()
```

```python
sns.lmplot(x="acceleration", y="mpg", hue = 'model_year', data=tips)
```
