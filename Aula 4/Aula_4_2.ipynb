```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style="darkgrid")
```

```python
df = sns.load_dataset('mpg')
df.head()
```

```python
sns.jointplot(x="mpg", y="cylinders", kind = 'reg', data=df)
```

```python
sns.jointplot(x="displacement", y="horsepower", kind = 'hex', color = 'black', data=df)
```

```python
sns.jointplot(x="weight", y="acceleration", kind = 'kde', data=df)
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, kind = 'reg')
```
