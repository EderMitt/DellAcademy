```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style="darkgrid")
```

```python
room = pd.read_csv('dados3.txt')
room.head()
```

```python
plt.figure(figsize=(10,5))
plt.title('Vendas ao longo dos meses')
sns.scatterplot(x="Month", y="Sales", data=room)
plt.show()
```

```python
plt.figure(figsize=(25,5))
plt.title('Vendas ao longo dos meses')
sns.lineplot(x="Month", y="Sales", data=room)
plt.show()
```
