```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style="darkgrid")
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
plt.figure(figsize=(10,5))
sns.stripplot(x="fare", y="embark_town", data=df)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.swarmplot(x="fare", y="embark_town", data=df)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.swarmplot(x="fare", y="embark_town", hue = 'sex', data=df)
plt.show()
```

```python
sns.catplot(x="fare", y="embark_town", hue = 'sex', col = 'survived', kind = 'swarm', data=df)
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="fare", y="embark_town", hue = 'sex', data=df)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.title('Idade de passageiros em cada classe por sexo')
sns.boxplot(x="fare", y="embark_town", hue = 'sex', data=df)
plt.show()
```

```python
plt.figure(figsize=(15,8))
sns.boxenplot(x="fare", y="embark_town", hue = 'sex', data=df)
plt.show()
```

```python
sns.catplot(x="fare", y="embark_town", hue = 'sex', col = 'survived', kind = 'boxen', data=df)
```

```python
plt.figure(figsize=(8,5))
sns.violinplot(x="fare", y="embark_town", hue = 'sex', data=df)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.violinplot(x="fare", y="embark_town", data=df)
sns.swarmplot(x="fare", y="embark_town",color = 'black', size = 3, data=df)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.violinplot(x="fare", y="embark_town", hue = 'sex', split = True, data=df)
plt.show()
```

```python
sns.catplot(x="fare", y="embark_town", hue = 'sex', col = 'survived', kind = 'violin', split = True, data=df)
```
